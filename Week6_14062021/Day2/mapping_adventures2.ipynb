```python
import pandas as pd
import plotly.express as px
import os
from pathlib import Path
from dotenv import load_dotenv
```

### Prep Mapbox API Credentials

```python
# Set up API credentials
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("MAPBOX_API_KEY")

# Set the Mapbox API
px.set_mapbox_access_token(map_box_api)
```

### Read in data

```python
# Read in data
places_of_interest = pd.read_csv(r"C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 6\2\08-Stu_Mapping_Adventures\Resources\nyc_places_interest.csv").dropna()

places_of_interest.head()
```

### Plot Data

#### Plot Parks that are of Interest

```python
# Slice and plot data by place type of park 
df_park = places_of_interest.loc[places_of_interest["PlaceType"] == 'Park']
df_park.head()
```

```python
plot1 = px.scatter_mapbox(df_park,
                          lat='Latitude',
                          lon='Longitude',
                          color='Name',
                          size='Id')
plot1.show()
```

#### Plot Gardens of Interest

```python
# Slice and plot data by place type of garden 
df_garden = places_of_interest.loc[places_of_interest["PlaceType"] == 'Garden']
df_garden.head()
```

```python
plot2 = px.scatter_mapbox(df_garden,
                          lat='Longitude',
                          lon='Longitude',
                          color='Name',
                          size='Id')
plot2.show()
```

#### Plot Squares of Interest

```python
# Slice and plot data by place type of square 
df_square = places_of_interest.loc[places_of_interest["PlaceType"] == 'Square']
df_square.head()
```

```python
plot3 = px.scatter_mapbox(df_square,
                          lat='Latitude',
                          lon='Longitude',
                          color='Name',
                          size='Id')
plot3.show()
```

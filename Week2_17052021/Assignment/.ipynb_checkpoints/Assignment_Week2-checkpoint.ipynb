```python
# Set up the same process to read in `sales_data.csv`. However, instead append every row of the sales data to a new `sales` list object.

#Read in `menu_data.csv` 
sales_data = Path(r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Homework\Week 2\Instructions\PyRamen\Resources\sales_data.csv')

#Initialize an empty `sales` list object to hold the contents of `sales_data.csv`.
sales = []

# Use a `with` statement and open the `sales_data.csv` by using its file path.
with open(sales_data, 'r') as sales_data_file:
    
    #Use the `reader` function from the `csv` library to begin reading `sales_data.csv`.  
    sales_data_reader = csv.reader(sales_data_file, delimiter=',')
    
    #Use the `next` function to skip the header (first row of the CSV).
    sales_data_header = next(sales_data_reader)
    print(sales_data_header)
    print()
    
    # Initialize an empty `report` dictionary to hold the future aggregated per-product results. The `report` dictionary will eventually contain the following metrics:
    report = {}
    
    #Loop over the rest of the rows and append every row to the `sales` list object (the outcome will be a list of lists).
    for row in sales_data_reader:
        #print(row)
        sales.append(row)            
        
        #For each row of the `sales` data, set the following columns of the sales data to their own variables
        [Quantity, Menu_Item] = row[-2], row[-1]

        count = 0 #the total quantity for each ramen type
        revenue = 0  #the total revenue for each ramen type
        cogs = 0   #the total cost of goods sold for each ramen type
        profit = 0  #the total profit for each ramen type
                   
        name = row[-1]        
       
        if name not in report.keys():
            report[name] = {'01-count': count, '02-revenue': revenue, '03-cogs': cogs, '04-profit': profit}
        else:
            report[name]['01-count'] += count
            report[name]['02-revenue'] += revenue
            report[name]['03-cogs'] += cogs
            report[name]['04-profit'] += profit        
          
       
print(f"Line_Item_ID: {sales[0][0]}")
print()
print(f"Date: {sales[0][1]}")
print()
print(f"Credit card number: {sales[0][2]}")                    
print()
print(f"Quantity: {sales[0][3]}")
print()
print(f"Menu item: {sales[0][4]}")
print()
print(report)
print()
```

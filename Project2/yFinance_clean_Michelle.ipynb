# Get stock market data & clean data

```python
%matplotlib inline
import yfinance as yf
import quantstats as qs
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

# extend pandas functionality with metrics, etc.
qs.extend_pandas()
```

```python
# Names of intersted stocks
stock_names = {'BHP.AX',
               'FMG.AX',
               'ABBV',
               'PFE',
               'BAC',
               'COKE',
               'SQ',
               'AAPL',
               'AMD',
               'FDX',
               'EBAY',
               'AMZN',
               'C',
               'F',
               'GDX',
               'USO'}
```

```python
# Get a list of all closing prices
stock_list = {}

for stock in stock_names:
    
    stock_list[stock] = qs.utils.download_returns(stock)


    
# Limit all close prices to the last 10 years 
close_all = []

for name in stock_names:
    
    close_new = stock_list[name].loc["2016-01-01":]
    
    close_all.append(close_new)
    
    
# Convert close prices to dataframe
df = pd.DataFrame(close_all[:]).T
df.columns = stock_names

display(df)
```

```python
#Check dataframe 
df.describe()
```

```python
#Drop duplicates
df.drop_duplicates(inplace=False)

df.head()
```

```python
# Count nulls
df.isnull().sum()
```

```python
# Drop NA
df = df.dropna()
```

```python
# final df
df.describe()
```

```python
# Count nulls
df.isnull().sum()
```

```python
#Percentage change
pct=df.pct_change()
pct.head()
```

```python
#Drop duplicates
pct.drop_duplicates(inplace=False)

pct.head()
```

```python
# Count nulls
pct.isnull().sum()
```

```python
# Drop NA
pct_final = pct.dropna()
pct_final
```

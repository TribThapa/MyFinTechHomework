### Import the necessary libraries and dependencies.

```python
# Import libraries and dependencies
from pathlib import Path
import csv
import pandas as pd
```

### Read CSV in as DataFrame

```python
# Set the path
data = Path("../Resources/loans.csv")

# Read in the CSV as a DataFrame
data_df = pd.read_csv(data)
data_df.head()
```

### Show the First 10 Records

```python
# Retrieve rows with index 0 up to 10 (not including)
data_df.head(10)
```

### View Summary Statistics

```python
# Describe summary statistics for csv data
data_df.describe()
```

```python
data_df.info()
```

### Filter Columns and Create Subset DataFrame

```python
# Filter the DataFrame down to the following, keep all rows

col = ['loan_amnt', 'term', 'int_rate', 'emp_title', 'annual_inc', 'purpose']

data_df_filtered = data_df.loc[:, col]
data_df_filtered

#loan_amnt
#term
#int_rate
#emp_title
#annual_inc
#purpose
```

### Filter Rows by 36 Month Terms

```python
# Conditional indexing to filter DataFrame where 'term' is equal to '36 months'
data_df_filtered_months = data_df_filtered.loc[data_df_filtered['term'] == '36 months']
data_df_filtered_months
```

### Modify Row Values

```python
# Change row values within the 'term' column from '36 months' to '3 Years'
data_df_filtered_months.iloc[:, 1] = '3 Years'
data_df_filtered_months
```

```python
# Change row values within the 'emp_title' column from NaN to 'Unknown'
data_df_filtered_months['emp_title'].fillna('Unknown', inplace=True)
data_df_filtered_months
```

```python
data_df_filtered_months['is_Null'] = data_df_filtered_months['emp_title'].isnull()

data_df_filtered_months.loc[data_df_filtered_months.is_Null == True, 'emp_title'] = "Unknown"

data_df_filtered_months.head()
```

### Show Summary Statistics for 3 Year Loans of Customers with Annual Income Greater Than $80,000

```python
# Display summary statistics where annual income is greater than $80,000 to find count and mean

data_df_filtered_months['annual_inc'] = data_df_filtered_months['annual_inc'].replace('RENT', '0')
data_df_filtered_months['annual_inc'] = data_df_filtered_months['annual_inc'].replace('MORTGAGE', '0')

data_df_filtered_months['annual_inc'] = data_df_filtered_months["annual_inc"].astype('float')

data_df_filtered_months_80000 = data_df_filtered_months.loc[data_df_filtered_months['annual_inc'] > 80000]

# Display summary statistics where annual income is less than $80,000 to find count and mean
data_df_filtered_months_80000.describe()
```

````python
# Create a DataFrame from the CSV file. Display the DataFrame's head
data_df = pd.read_csv(data)
data_df.head()
````

### Replace Columns

````python
# Display the current column names
data_df.columns
````

````python
# Set the new names for the columns. Replace the current names of the columns with the new names. Display the DataFrame's head.
col_names = ['Full Name', 'Email address', 'Postal Address', 'Zip code', 'Credi card number', 'Total balance']

data_df.columns = col_names
data_df.head()
````

### Rename Columns

The `rename` method can be used to replace or update column names selectively. Simply provide a dictionary to the column's parameter that has the following format:

```python
{
"Old Column Name": "New Column Name"
}
```

````python
# Use the `rename` method to change the name of the columns. Display the DataFrame's head.
data_df = data_df.rename(columns={
    "Full Name": "full_name",
    "Email address": "email@"
})

data_df.head()
````

### Reorder Columns

````python
# Reorder the columns by creating a new DataFrame. Display the DataFrame's head.
data_df = data_df[['Total balance', 'Postal Address','email@', 'Zip code', 'full_name', 'Credi card number', ]]

data_df.head()
````

### Create Columns

````python
# Add a new column to the DataFrame. Display the DataFrame's head.
data_df['Balance (1k)'] = data_df["Total balance"]/1000
data_df.head()
````

### Split Columns

````python
# Create a new DataFrame by splitting the text content of a column
names = data_df["full_name"].str.split(" ", expand=True)
names.head()
````

````python
PostalAddress = data_df["Postal Address"].str.split(" ", expand=True)
PostalAddress.head()
````

````python
# Add two new columns to the `customer_dataframe` using the columns created after splitting the `full_name` column
data_df['first name'] = names[0]
data_df['last name'] = names[1]
data_df.head()
````

### Delete Columns

````python
# Use the `drop` method to delete a column from the `customer_dataframe`
data_df = data_df.drop(columns=['full_name'])
data_df.head()
````

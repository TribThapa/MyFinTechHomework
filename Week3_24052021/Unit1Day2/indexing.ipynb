### Index Selection Using Loc

```python
people_csv.info()
```

```python
# Set the index as the 'first_name' column
people_csv_index = people_csv.set_index('first_name')
people_csv_index
```

```python
# Sort the index
people_csv_index_sort = people_csv_index.sort_index()
people_csv_index_sort
```

```python
# Select the row with the index 'Evan'. lod: label based indexgin, iloc: integer based indexing.
i = people_csv_index_sort.loc['Evan']
i
```

```python
# Slice the data to output a range of rows based on the index
j = people_csv_index_sort.loc[['Evan'], :]
j
```

```python
# Filter rows based on a column value conditional
people_csv_index_sort.loc[people_csv_index_sort['age'] < 50]
```

```python
# Create a list of groceries
Groceries = ["Tomatoes", "Potatoes", "Onions", "Garlic", "Ginger", "Cauliflower", "Broccoli", "SpringOnions", "Chilli", "Coriander"]

# Find the first two items on the list
print("The first two elements on the groceries list are: ")
print(Groceries[: 2])
print()

# Find the last five items on the list
print("These are the last five items on the groceries list:")
print(Groceries[-5:])
print()

# Find every other item on the list, starting from the second item
print("These are every other item from the groceries list starting with the second item:")
print(Groceries[1 :: 2])
print()

# Find every other item on the list, starting from the second item
print("These are every other item from the groceries list starting with the first item:")
print(Groceries[0 :: 2])
print()

# Find every other item on the list, starting from the second item
print("These are every third item from the groceries list starting with the first item:")
print(Groceries[0 :: 3])
print()

# Add an element to the end of the list
print("Adding 'Cucumber' to the groceries list....")
Groceries.append("Cucumber")
print(Groceries)
print()

# Removing a specified element within the list at the given index
print("Changing a specified elemnt within the groceries list at a given index")
Potatoes_Index = Groceries.index("Potatoes")
print(Potatoes_Index)
Groceries.pop(Potatoes_Index)
print(Groceries)
print()

# Change a specified element within the list at the given index
print("Changing a specified elemnt within the groceries list at a given index")
Tomatoes_Index = Groceries.index("Tomatoes")
print(Tomatoes_Index)
Groceries[0]= "Beans"
print(Groceries)
print()

# Calculate how many items you have in the list
print("Calculating the number of items in the groceries list.....")
a = len(Groceries)
print(f"There are {a} items on the groceries list")
print()

# Find the index of the particular element name
print("Finding the index of an element.....")
SpringOnions_Index = Groceries.index("SpringOnions")
print(f"The index for Spring onions is {SpringOnions_Index}")
print()

# Remove an element from the list based on the given element name
print("Removing 'Chilli' element based upon its name.....")
Groceries.remove("Chilli")
print(Groceries)
print()

# Remove an element from the list based on the given index
print("Removing 'Ginger' element based upon its index.....")
print(Groceries)
GingerIndex = Groceries.index("Ginger")
print(GingerIndex)
Groceries.pop(GingerIndex)
print(Groceries)
print()

# Remove the last element of the list
print("Removing the last element of the list.....")
print(Groceries)
Groceries.remove(Groceries[-1])
print(Groceries)
print()
```

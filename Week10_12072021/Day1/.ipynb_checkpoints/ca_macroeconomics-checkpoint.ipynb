# CA Macroeconomics

```python
# Import libraries and dependencies
import pandas as pd
import datetime
import statsmodels.api as sm
%matplotlib inline
```

## Hodrick-Prescott filter analysis

### GDP Analysis

```python
# Read the GDP csv data and set the column `DATE` as a datetime index
gdp_df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\1\05-Stu_Hodrick_Prescott\Resources\ca_gdp_data.csv'),
                     parse_dates=True,
                     index_col='DATE')

gdp_df.head()
```

```python
# Plot the raw GDP data as a line plot
gdp_df.GDP.plot(title='Raw GDP')
```

```python
# Use the `hpfilter` function to decompose the GDP column into the trend and noise components.
gdp_noise, gdp_trend = sm.tsa.filters.hpfilter(gdp_df["GDP"])
```

```python
# Plot the GDP trend data as a line plot
gdp_trend.plot(title='GDP trend')
```

```python
# Plot the GDP noise data as a line plot
gdp_noise.plot(title='GDP noise')
```

### Inflation

```python
# Read the Inflation csv data and set the column `DATE` as a datetime index
# Note: Inflation, as measured by consumer price index in urban areas
inflation_df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\1\05-Stu_Hodrick_Prescott\Resources\ca_inflation.csv'),
                           parse_dates=True,
                           index_col='DATE')

inflation_df.head()
```

```python
# Plot the raw inflation data as a line plot
inflation_df.INFLATION_RATE.plot(title='Raw Inflation Rate')
```

```python
# Use the `hpfilter` function to decompose the `INFLATION_RATE` column into the trend and noise components.
inflation_noise, inflation_trend = sm.tsa.filters.hpfilter(inflation_df["INFLATION_RATE"])
```

```python
# Plot the inflation trend data as a line plot
inflation_trend.plot(title='Inflation rate trend')
```

```python
# Plot the inflation noise data as a line plot
inflation_noise.plot(title='Inflation rate noise')
```

```python
# Plot the exponentially-weighted moving average for the inflation data
inflation_df.INFLATION_RATE.ewm(halflife=10).mean().plot(title='Exponentially weighted moving average')
```

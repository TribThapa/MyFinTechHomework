```python
from nltk.corpus import reuters, stopwords
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

lemmatizer = WordNetLemmatizer()
```

```python
article = reuters.raw(reuters.fileids(categories='cpi')[2])
print(article)
```

```python
def process_text(doc):
    
    sw = set(stopwords.words('english'))
    
    regex = re.compile("[^a-zA-Z ]")
    
    re_clean = regex.sub('', article)
    
    words = word_tokenize(re_clean)
    
    lem = [lemmatizer.lemmatize(word) for word in words]
    
    output = [word.lower() for word in lem if word.lower() not in sw]
    
    return output
```

```python
processed = process_text(article)

print(processed)
print()
print(len(processed))
```

# Task 4

```python
# Corpus - list of articles about grains
ids = reuters.fileids(categories='grain')

corpus = [reuters.raw(i) for i in ids]
```

```python
# Define preprocess function
def process_text(doc):    
    
    sw = set(stopwords.words('english'))
    
    regex = re.compile("[^a-zA-Z ]")
    
    re_clean = regex.sub('', doc)
    
    words = word_tokenize(re_clean)
    
    lem = [lemmatizer.lemmatize(word) for word in words]
    
    output = [word.lower() for word in lem if word.lower() not in sw]
    
    return output
```

```python
# Define the counter function
def word_counter(corpus):
    
    string = ' '.join(corpus)
    
    preprocess = process_text(string)
    
    word_counts = Counter(preprocess)
    
    top_10 = dict(word_counts.most_common(20))
    
    return top_10
```

```python
final_result = word_counter(corpus)
```

```python
import pandas as pd

df = pd.DataFrame(list(final_result.items()),
                  columns=["Words", "Count"])

df_sorted = df.sort_values(by=["Count"], ascending=False)

df_sorted.head(10)
```

```python
# Define the counter function
def big_ram(corpus):
    
    string = ' '.join(corpus)
    
    preprocess = process_text(string)
    
    bigram_counts = Counter(ngrams(preprocess, n=2))
    
    top_20 = dict(bigram_counts.most_common(20))
    
    return top_20
```

```python
final_bigram = big_ram(corpus)
```

```python
bigram_df = pd.DataFrame(list(final_bigram.items()),
                         columns=["Words", "Count"])

bigram_df_sorted = bigram_df.sort_values(by=["Count"], ascending=False)

bigram_df_sorted.head(10)
```

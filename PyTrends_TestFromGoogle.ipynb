```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import os
import matplotlib.pyplot as plt
```

```python
#execute the TrendReq method by passing the host lanaguage(hl) and timezone(tz) parameters. https://medium.com/the-data-science-publication/how-to-use-the-pytrends-api-to-get-google-trends-data-4378acbaaa8a
pytrends = TrendReq(hl='en', tz=360)
```

```python
#build list of keywords 
kw_list = ['ai', 'chicken', 'space']

#build payload which contains relevant request information that helps the service(e.g., Google) understand what data you want to obtain 
pytrends.build_payload(kw_list, timeframe='2015-01-01 2021-01-01', geo='US')
```

```python
df = pytrends.interest_over_time()
print(df.head(20))
```

```python
a = df.sort_values(by='chicken', ascending=True)[0:30]
print(a.head(10))
```

```python
a.plot(kind='bar', figsize=(20,4))
```

```python
#Alpaca data
pytrends.build_payload(kw_list =['Alpaca'])
```

```python
interest_over_time_df = pytrends.interest_over_time()

print(interest_over_time_df.head())
```

```python
#Interest by region
interest_by_region_df = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)

print(interest_by_region_df.head())
```

```python
plt.figure()
interest_by_region_df.sort_values(by='Alpaca', ascending=False)[0:30].plot(kind='bar', figsize=(20,4), sort_columns=True)
```

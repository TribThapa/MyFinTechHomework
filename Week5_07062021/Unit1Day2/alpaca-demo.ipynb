# Alpaca Demo

```python
# Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Alpaca API key and secret. # Verify that Alpaca key and secret were correctly loaded.
alpaca_key = os.getenv("ALPACA_KEY")
print(alpaca_key)

alpaca_key_secret = os.getenv("ALPACA_KEY_SECRET")
print(alpaca_key_secret)
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(alpaca_key,
                       alpaca_key_secret, 
                       api_version=2)
```

```python
# Format current date as ISO format
today = pd.Timestamp("2020-06-04", tz="America/New_York").isoformat()
```

```python
# Set the tickers
tickers = ['GOOG', 'AAPL', 'FB']
```

```python
# Set timeframe to one day ('1D') for the Alpaca API
timeframe = "1D"
```

```python
# Get current closing prices for FB and TWTR
df_portfolio = alpaca.get_barset(tickers,
                                 timeframe=timeframe,
                                 start=today,
                                 end=today).df

# Display sample data
df_portfolio.head()
```

```python
# Format start and end dates as ISO format for one year period
start_date = pd.Timestamp("2019-06-04", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-06-04", tz="America/New_York").isoformat()
```

```python
# Get closing prices for FB and TWTR from the last year
df_portfolio_year = alpaca.get_barset(tickers, 
                                      timeframe=timeframe,
                                      start=start_date,
                                      end=end_date,
                                      limit=1000).df
# Display sample data
df_portfolio_year.head(10)
```

```python
# Create an empty DataFrame for closing prices
df_close_price = pd.DataFrame()

# Fetch the closing prices of FB and TWTR
df_close_price['AAPL'] = df_portfolio_year["AAPL"]['close']
df_close_price['FB'] = df_portfolio_year["FB"]['close']
df_close_price['GOOG'] = df_portfolio_year["GOOG"]['close']

# Drop the time component of the date
df_close_price.index = df_close_price.index.date

# Display sample data
df_close_price.head()
```

```python
# Compute daily returns
daily_returns = df_close_price.pct_change().dropna()

# Display sample data
daily_returns
```

```python
# Plot daily returns
daily_returns.plot(figsize=(15,10))
```

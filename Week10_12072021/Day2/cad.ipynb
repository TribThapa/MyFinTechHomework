```python
# Import libraries and dependencies
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Read CSV, parse dates, and set index
df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\2\08-Stu_CAD\Resources\euro_cad.csv'),
                 parse_dates=True, 
                 index_col='Date')
df.head(2)
```

```python
# Daily EUR to CAD exchange rates
df.Rate.plot()
```

```python
df["Rate"].plot()
```

```python
# Daily percentage changes in exchange rate (resampled to daily)
df["Return"] = df["Rate"].pct_change() * 100 * 24
df = df.resample('D').mean()
df = df.dropna()
df.head()
```

```python
# Plot returns
df["Return"].plot()
```

# GARCH

```python
# Create the GARCH model
from arch import arch_model
returns = df["Return"]

model = arch_model(returns,
                   mean="Zero",
                   vol="GARCH",
                   p=1,
                   q=1)

```

```python
#Fit the model
results = model.fit()
```

```python
# Output the model summary
results.summary()
```

```python
# Plot the model estimate of the annualized volatility
fig = results.plot(annualize='D')
```

# Forecasting from GARCH

```python
# Find the last day of the dataset
df.tail()
```

```python
# Create a 5 day forecast of EUR to CAD exchange rate volatility
forecast_horizon = 5

# Take the last day of the data we used above. 
# If forecast horizon is 3, then the resulting 'h.1', 'h.2', and 'h.3' 
# are the forecasts for the following 3 days.
forecasts = results.forecast(start="2019-08-23",
                            horizon=forecast_horizon)

forecasts
```

```python
# Annualize the forecast
annualised_forecast = np.sqrt(forecasts.variance.dropna() * 252)

annualised_forecast
```

# Visualizing the forecast

```python
# Transpose the forecast so that it is easier to plot
final_df = annualised_forecast.dropna().T
final_df
```

```python
# Plot the final forecast
final_df.plot()
```

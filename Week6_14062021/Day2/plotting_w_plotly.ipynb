```python
import plotly.express as px
import pandas as pd
from pathlib import Path
```

## Plot Foreclosures

```python
# Read in data
foreclosures = pd.read_csv((r"C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 6\2\02-Stu_Plotting_w_Plotly\Resources\allegheny_foreclosures1.csv"),
infer_datetime_format=True,
parse_dates=True,
index_col="filing_date")

foreclosures.head()
```

```python
# Slice data and group
municipality = foreclosures[["municipality"]]
amount = foreclosures[["amount"]]

# df_new = foreclosures[["municipality", "amount", "filing_date"]]
# df_new['year'] = pd.DatetimeIndex(foreclosures['filing_date']).year
# df_new.set_index('filing_date', inplace=True)
# df_new.sort_values(by=['municipality'], inplace=True)
# df_new.head()
```

```python
df_grp = foreclosures.groupby([foreclosures.index.year, 'municipality']).count()
df_grp.head()
```

```python
df_grp.reset_index(inplace=True)
df_grp.drop(['plaintiff'], axis=1, inplace=True)
df_grp.head()
```

### Plot relationship between municipality and the number of foreclosures

```python
# Create scatter plot
plot1 = px.scatter(df_grp, x="municipality", y="amount", color='filing_date')
plot1
```

### Plot the relationship between filing date and municipality

```python
# Create scatter plot
plot2 = px.scatter(df_grp, x='municipality', y='filing_date', size='amount') 
plot2
```

```python
# Import libraries and dependencies
import pandas as pd
%matplotlib inline
```

```python
# Read in CSV, parse dates, and set index
df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\1\04-Ins_Hodrick_Prescott\Resources\IVV.csv'),
                 parse_dates=True, 
                 index_col='date')

# Display bottom rows
df.tail()
```

```python
df["close"].plot()
```

```python
# Moving average. Each value is the average of previous 10 values.
df["close"].rollingling(window=10).mean().plot()
```

```python
# Exponentially-Weighted Moving Average
df["close"].ewm(halflife=10).mean().plot()
```

```python
#Hodrick-Prescott filter
import statsmodels.api as sm

ts_noise, ts_trend = sm.tsa.filters.hpfilter(df["close"])
```

```python
ts_noise.plot()
```

```python
ts_trend.plot()
```

```python
# Read the GDP csv data and set the column `DATE` as a datetime index
gdp_df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\1\05-Stu_Hodrick_Prescott\Resources\ca_gdp_data.csv'),
                     parse_dates=True,
                     index_col='DATE')

gdp_df.head()
```

```python
#Plot the raw GDP data as a line plot.
gdp_df["GDP"].plot()
```

```python
# Use the `hpfilter` function to decompose the GDP column into the trend and noise components.
gdp_noise, gdp_trend = sm.tsa.filters.hpfilter(gdp_df["GDP"])
```

```python
# Plot the GDP trend data as a line plot.
gdp_trend.plot.line()
```

```python
# Plot the GDP noise as a line plot.
gdp_noise.plot.line()
```

```python
# Read the Inflation csv data and set the column `DATE` as a datetime index
# Note: Inflation, as measured by consumer price index in urban areas
inflation_df = pd.read_csv((r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 10\1\05-Stu_Hodrick_Prescott\Resources\ca_inflation.csv'),
                           parse_dates=True,
                           index_col='DATE')

inflation_df.head()
```

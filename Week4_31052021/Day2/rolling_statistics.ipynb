# Visualizing Rolling Statistics with Pandas DataFrames


## 1. Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path

%matplotlib inline
```

## 2. Create a Path to the File Using Pathlib.

```python
# Read the MSFT data, set the `date` as the index
MSFT_data = r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 4\2\09-rolling-statistics-01\Resources\MSFT.csv'
```

## 3. Read the CSV into a Pandas DataFrame.

```python
# Read the MSFT CSV into a DataFrame while formatting and setting the `date` column as the index
MSFT_data_df = pd.read_csv(MSFT_data, infer_datetime_format=True, parse_dates=True, index_col='date')

MSFT_data_df.head()
```

## 4. Drop Extra Columns

```python
# Use the `drop` function to drop the extra columns
MSFT_data_drop_df = MSFT_data_df.drop(columns=['volume', 'open', 'high', 'low'])
MSFT_data_drop_df.head()
```

## 5. Plot Daily Closing Price of MSFT

```python
# Plot the daily closing prices of MSFT
plt1 = MSFT_data_drop_df.plot()
plt1.legend(['daily close'])
```

## 6. Rolling 7-Day Mean of MSFT Closing Prices

```python
# Plot a rolling 7-day mean of MSFT's closing prices
plt2 = MSFT_data_drop_df.rolling(window=7).mean().plot()
plt2.legend(['Rolling 7-day mean'])
```

## 7. Rolling 7-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 7-day std dev of MSFT's closing prices
plt3 = MSFT_data_drop_df.rolling(window=7).std().plot()
plt3.legend(['Rolling 7-day std'])
```

## 8. Rolling 30-Day Mean of MSFT Closing Prices

```python
# Plot a rolling 30-day mean of MSFT's closing prices
plt4 = MSFT_data_drop_df.rolling(window=30).mean().plot()
plt4.legend(['Rolling 30-day mean'])
```

## 9. Rolling 30-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 30-day std dev of MSFT's closing prices
plt5 = MSFT_data_drop_df.rolling(window=30).std().plot()
plt5.legend(['Rolling 30-day std'])
```

## 10. Rolling 180-Day Mean of MSFT Closing Prices

```python
# Plot a rolling 180-day mean of MSFT's closing prices
plt6 = MSFT_data_drop_df.rolling(window=180).mean().plot()
plt6.legend(['Rolling 180-day mean'])
```

## 11. Rolling 180-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 180-day std dev of MSFTs closing prices
plt7 = MSFT_data_drop_df.rolling(window=180).std().plot()
plt7.legend(['Rolling 180-day std'])
```

## 12. Overlay Daily Closing Price and 180-Day Rolling Mean of MSFT

```python
# Set figure of the daily closing prices of MSFT
Fig1 = MSFT_data_drop_df.plot(figsize=(15, 10), color='k')

# Plot 180-Day Rolling Mean on the same figure
MSFT_data_drop_df.rolling(window=180).mean().plot(ax=Fig1, color='r', alpha=0.3)

# Set the legend of the figure
Fig1.legend(['daily close', '180-day rolling mean'])
```

## Ice Breakers on Request

### Import requests library

```python
import requests
import json
```

```python
url = 'http://api.worldbank.org/v2/country/ca/indicator/NY.GDP.MKTP.CD?format=json'

response = requests.get(url)

print(response.status_code)
```

```python
response_content = response.content
print(response_content)
```

```python
data = response.json()
print(data)
```

```python
data_json = json.dumps(data, indent=4, sort_keys=True)
print(data_json)
```

```python
response = requests.get(url).json()

response_json = json.dumps(response, indent=4, sort_keys=True)
print(response_json)
```

```python
country = data[1][1]['country']['value']
gdp_value = data[1][1]['value']

print(f'Country {country}')
print(f'Value: {gdp_value}')
```

### Declare `request_urls`

```python
# Declare `url` variables
prog_joke_url = "https://official-joke-api.appspot.com/jokes/programming/random"
rand_joke_url = "https://official-joke-api.appspot.com/jokes/random"
cat_facts_url = "https://catfact.ninja/facts"
daily_releases_url = "https://www150.statcan.gc.ca/n1/dai-quo/ssi/homepage/daily-banner-eng.json"
us_gdp_data_url = "http://api.worldbank.org/v2/country/us?format=json"
```

### Execute `GET` request using `requests` library

```python
# Execute `GET` request with url
prog_req = requests.get(prog_joke_url)
rand_req = requests.get(rand_joke_url)
cat_req = requests.get(rand_joke_url)
daily_req = requests.get(daily_releases_url)
us_gdp_req = requests.get(us_gdp_data_url)
```

### Get response `status code`

```python
# Print `response_data variable` 200: implies your request was successful and the server responded with the data you were requesting, whereas 404 implies an unsuccessful request.
prog_req.status_code
```

```python
rand_req.status_code
```

```python
cat_req.status_code
```

```python
daily_req.status_code
```

```python
us_gdp_req.status_code
```

### Extract response `content`

```python
# Store response using `content` attribute
prog_req_content = prog_req.content
print(prog_req_content)
```

```python
rand_req_content = rand_req.content
print(rand_req_content)
```

### Convert output to JSON

```python
prog_json = prog_req.json()
print(prog_json)
```

```python
rand_json = rand_req.json()
print(rand_json)
```

```python
cat_json = cat_req.json()
print(cat_json)
```

### Format data with `json.dumps`

```python
# Use json.dumps to format data
prog_json_dmps = json.dumps(prog_json, indent=4, sort_keys=True)
print(prog_json_dmps)
```

```python
rand_json_dmps = json.dumps(rand_json, indent=4, sort_keys=True)
print(rand_json_dmps)
```

```python
cat_json_dmps = json.dumps(cat_json, indent=4, sort_keys=True)
print(cat_json_dmps)
```

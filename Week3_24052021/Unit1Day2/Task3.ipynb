### CHANGE COLUMN NAME

```python
ColName = ['Full Name', 'Email', 'Address', 'Zip Code', 'Credit Card Number', 'Account Balance']

data_df.columns = ColName
data_df.head(10)
```

### RENAME COLUMNS

```python
data_df = data_df.rename(columns={
    'Full Name': 'full_name',
    'Credit Card Number': 'credit_card_number'
})

data_df.head(10)
```

### REORDER COLUMNS

```python
data_df = data_df[['credit_card_number', 'Account Balance', 'full_name', 'Email', 'Address', 'Zip Code']]

data_df.head(10)
```

### CREATE AND ADD NEW COLUMN

```python
data_df['Balance (1k)'] = data_df['Account Balance'] / 1000

data_df.head(10)
```

### SPLIT COLUMNS AND ADD TO DATAFRAME

```python
names = data_df['full_name'].str.split(" ", expand=True)
names.head()
```

### REARRANGED DATAFRAME

```python
data_df['first_name'] = names[0]
data_df['last_name'] = names[1]

data_df = data_df[['first_name', 'last_name', 'full_name', 'Email', 'Address', 'Zip Code', 'credit_card_number', 'Account Balance', 'Balance (1k)']]
data_df.head(5)
```

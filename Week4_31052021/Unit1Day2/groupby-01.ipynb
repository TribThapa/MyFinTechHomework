# Grouping Data in Pandas DataFrames

## 1. Import Libraries and Dependencies

```python
# Import necessary libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

## 2. Create a Path to the File Using Pathlib

```python
# Use the Pathlib libary to set the path to the CSV
data = r'C:\Users\TribThapa\Desktop\Thapa\ResearchFellow\Courses\FinTech_Bootcamp_MonashUni2021\monu-mel-virt-fin-pt-05-2021-u-c\Activities\Week 4\2\06-groupby-01\Resources\people_cleansed.csv'
```

## 3. Read the CSV into a Pandas DataFrame

```python
# Use the file path to read the CSV into a DataFrame and display a few rows
data_df = pd.read_csv(data)
data_df.head()
```

## 4. Group DataFrame by `Occupation` and perform `count` aggregation

```python
# Group by `Occupation` and perform count
data_grp_df = data_df.groupby(['Occupation']).count()
data_grp_df
```

## 5. Group DataFrame by `Occupation` and Calculate Average Salary and Age

```python
# Calculate average Salary and Age for each Occupation
data_grp_df = data_df.groupby(['Occupation'])['Person_ID', 'Salary', 'Age'].mean()
data_grp_df.head(10)
```

## 6. Group DataFrame by `Occupation` **and** `Gender` then Calculate Average Salary and Age

```python
# Group by more than one column
data_grp_df = data_df.groupby(['Occupation', 'Gender'])['Salary', 'Age'].mean()
data_grp_df.head()
```
